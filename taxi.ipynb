```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("taxis")

df
```

```python
# 1.1) Xác định 5 khung giờ cao điểm (nhiều chuyến xe nhất)
df.groupby(df["pickup"].dt.hour)["pickup"].count().nlargest(5)
```

```python
# 1.2) Tạo cột same_borough có giá trị là 1 nếu cùng điểm pickup và dropoff, và nhận giá trị 0 nếu ngược lại
df["same_borough"] = (df["pickup_borough"] == df["dropoff_borough"]).astype(int)
```

```python
df[["pickup_borough", "dropoff_borough", "same_borough"]].head(5)
```

```python
# 1.3) Đếm số chuyến xe có điểm pickup và dropoff a) cùng borough và b) khác borough
df["same_borough"].value_counts()
```

Bai 2:

```python
# 2.1) Tính trung bình quãng đường di chuyển cho các chuyến xe xuất phát từ Manhattan
cond = df["pickup_borough"] == "Manhattan"

df.loc[cond, "distance"].mean()
```

```python
# 2.2) Tạo cột duration là thời gian giữa pickup và dropoff theo phút (Hint: tìm hiểu .dt.total_seconds() của time delta)
s = df["dropoff"] - df["pickup"]

df["duration"] = s.dt.total_seconds() / 60
df["duration"]
```

```python
# 2.3) Tính trung bình thời gian di chuyển cho mỗi chuyến xe có điểm pickup và dropoff 
# a) cùng borough và 
# b) khác borough. Bạn thấy kết quả tính được có hợp lý?

df.groupby("same_borough")["duration"].mean()
```

```python
# 2.4) Tạo cột pickup_hour là khung giờ pickup (Ví dụ từ 8:00:00 đến 8:59:5999999) sẽ có giá trị là 8
df["pickup_hour"] = df["pickup"].dt.hour
```

```python
# 2.5) Tạo cột km_per_min là số km đi được mỗi phút (Hint: số km lấy từ cột distance , số phút bạn đã tính trước đó)

df["km_per_min"] = df["distance"] / df["duration"]
```

```python
df[["km_per_min", "distance", "duration"]]
```

```python
# 2.6) Tính tốc độ trung bình cho chuyến xe trong mỗi khung giờ pickup. Sắp xếp theo chiều tăng dần của khung giờ

df.groupby("pickup_hour")["km_per_min"].mean().sort_index()
```
